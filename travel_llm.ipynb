```python
from groq import Groq
import os
from dotenv import load_dotenv

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

client = Groq(api_key=api_key)

def ask_travel_bot(question):
    system_prompt = """
    You are a professional travel planner AI.
    Give a simple, clear 1-day itinerary in bullet points.
    Include places, food, budget, and tips.
    """
    
    chat = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": question}
        ]
    )
    
    # Access the content attribute instead of using dictionary indexing
    return chat.choices[0].message.content

print(ask_travel_bot("Plan a 1-day trip in Mysore for a student with low budget."))
```
